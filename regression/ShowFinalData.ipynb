```python
import os
os.listdir("./finalData/")
```

```python
results = []
for i in os.listdir("./finalData/"):
    if i==".ipynb_checkpoints":
        continue
    df = pd.read_csv("./finalData/"+i)
    result = {}
    result["name"] = i.split("-")[0]
    result["type"] = i.split("-")[2].split(".")[0]
    result["Accuracy"] = (df["TP"].mean()+df["TN"].mean())/(df["TP"].mean()+df["TN"].mean()+df["FP"].mean()+df["FN"].mean())
    result["Precision"] = (df["TP"].mean())/(df["TP"].mean()+df["FP"].mean())
    result["Recall"] = (df["TP"].mean())/(df["TP"].mean()+df["FN"].mean())
    result["Specificity"] = (df["TN"].mean())/(df["TN"].mean()+df["FP"].mean())
    result["NPV"] = (df["TN"].mean())/(df["TN"].mean()+df["FN"].mean())
    results.append(result)
    
dfResults = pd.DataFrame(results)
dfResults
```
